```python

#Environment Building

print("   Welcome to vaccum cleaning agent    \n")
n=0
m=0
def create_envi(n, m):
    envi = []
    for a in range(n):
        row = []
        for b in range(m):
            value = 1 if random.random() < 0.06 else 0
            row.append(value)
        envi.append(row)
    return envi
def print_envi(envi):
    for val in envi:
      for f in val:
        print("[",f,"]", end=" ")
      print()
z=0
while(z<10):
  if(8<=n<=15 and 8<=m<=15):

    #Environment printing

    z=10
    print("The chosen environment is:")
    envi = create_envi(n, m)
    print_envi(envi)
    print("# 1 refers to rooms with dirt\n# 0 refers to rooms that are clean")
  else:
    z=0
    print("\nchoose the dimentions of the environment\n")
    n = int(input("Choose nth dimention(8 to 15): "))
    m = int(input("Choose mth dimention(8 to 15): "))
```

```python
#Agent development
def agent(status,location):
  if (status == 1):
    envi[location[0]][location[1]]=0
    print("Room [",location[0],",",location[1],"] is cleaned")
print("Agent is ready")
```

```python
#Cleaning and moving
location =[0,0]
def move(location):
  for k in range(0,n):
    for l in range(0,m):
      location = [k,l]
      #print(location[0])
      #print(location[1])
      status = envi[location[0]][location[1]]
      agent(status,location)
move(location)
```

#Agent based on an 8hr shift

```python
##choosing the environment
print("   Welcome to vaccum cleaning agent    \n")
o=0
p=0
z=10
n = 0
m = 0
point = 0
def choose(o,p):
  print("\nchoose the dimentions of the environment\n")
  o = int(input("Choose nth dimention(8 to 15): "))
  p = int(input("Choose mth dimention(8 to 15): "))
  return o,p
n,m = choose(o,p)
##8Hrs work shift

while(z==10):

  #Agent code
  if(8<=n<=15 and 8<=m<=15 ):
    z=0
    for hrs in range(0,8):
      #Environment Build
      def create_envi(n, m):
          envi = []
          for a in range(n):
              row = []
              for b in range(m):
                  value = 1 if random.random() < 0.06 else 0
                  row.append(value)
              envi.append(row)
          return envi
      def print_envi(envi):
          for val in envi:
            for f in val:
              print("[",f,"]", end=" ")
            print()

      print("The environment is:")
      envi = create_envi(n, m)
      print_envi(envi)
      print("# 1 refers to rooms with dirt\n# 0 refers to rooms that are clean")

      #Agent
      def agent(status,location):
        if (status == 1):
          envi[location[0]][location[1]]=0
          print("Tile [",location[0],",",location[1],"] is cleaned")

      print("\nAgent got ready\n")

      #points
      def points(envi,point):
        for cars in envi:
          for bikes in cars:
            if bikes==1:
              point = point + 1
        return point
      points(envi,point)
      point = points(envi,point)

      #Cleaning and moving
      location =[0,0]
      def move(location):
        for k in range(0,n):
          print("Moved Down to [",k,",","0 ] tile")
          for l in range(0,m):
            location = [k,l]
            if(l!=0):
              print("Moved right to [",k,",",l,"] tile")
            #print(location[0])
            #print(location[1])
            status = envi[location[0]][location[1]]
            agent(status,location)
      move(location)


      #After cleaning
      print("\n")
      print("Room after cleaning:\n")
      print_envi(envi)
      if(hrs<8):
        print("\n\n______________",hrs+1,"th HOUR DONE""____________\n\n")
        if(hrs<7):
          print("Next hour dirt has been added\n")
      if(hrs==7):
        print("XXXXXX---------SHIFT DONE---------XXXXXX")



  #Invalid n and m inputs
  else:
    z=10
    print("choose n and m in range")
    choose(o,p)
    n,m=choose(o,p)


```

#Agent's performance

```python
point = points(envi,point)
print("Number of points earned is:")
print(point)
num_of_rooms = n*m
count = 0
for i in range(0,8):
  for i in range (0,num_of_rooms):
    value = 1 if random.random() < 0.06 else 0
    if value==1:
      count = count + 1

# The number of tiles that have a chance of being dirty from nxm with 5% chance of each tile getting dirty is count

print("The total number of dirty tiles given (nXm) tiles and each having dirt being 5 percent chance is around:")
print(count)
if(point>count):
  print("My performance measure of this agent give n =",n,"and m=",m,"is:")
  print("100")
else:
  performance = (point/count)*100
  print("My performance measure of this agent give n =",n,"and m=",m,"is:")
  print(performance, "%")
```
